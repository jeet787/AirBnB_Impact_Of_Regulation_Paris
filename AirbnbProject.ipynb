<div style="background-color: #00A699; padding: 25px; border-radius: 10px;">
  <h1 style="color: #333; text-align: center"><strong>Finding average Price of AirBnB in Paris as per Localities</strong></h1>
</div>

```python
paris_listings_neighbourhood=paris_listings.groupby("neighbourhood").agg({"price":"mean"}).sort_values("price",ascending=False)
```

```python
paris_listings_neighbourhood
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.Figure(figsize = (12,8))
sns.barplot(data=paris_listings_neighbourhood,  x= "price" ,  y= paris_listings_neighbourhood.index , color="#00A699", palette="Set2")
plt.title("Neighbourhoods in paris by price")
plt.xlabel("How Expensive (Euros)")
plt.ylabel("Prominent Places with Air Bnb Located") 
plt.show()
```

 Inside of Paris , the price of Air BnB is directly proportional to popularity of the tourist place <br>
 or the view of Eiffel Tower.<br>
 We need to further see if the prices are impacted when the regulations are imposed.

```python
paris_listing_accomodates = paris_listings.groupby("accommodates").agg({"price":"mean"}).sort_values("price",ascending=False)
```

```python
paris_listing_accomodates
```

The Airbnbs withn 15 accomodates is the costliest , even costlier than 16 accomodates <br>
this might be due to luxury, or the view of prominent tourist spots like Eiffel Tower <br>
There might be outliers in the 15 accomodates, which is increasing its average price<br>

```python
plt.figure(figsize=(12,8))
sns.barplot(data=paris_listing_accomodates,  x= "price" ,y= paris_listing_accomodates.index , color="#00A699", palette="Set2",orient="horizontal",order=paris_listing_accomodates.index)
plt.title("Neighbourhoods in paris by price")
plt.xlabel("How Expensive (Euros)")
plt.ylabel("Prominent Places with Air Bnb Located") 
plt.show()
```

```python
paris_listings["year_of_listing"]=paris_listings["host_since"].dt.year #to create a new column year_of_listing by extracting the year from the host_since column
```

```python
paris_listings.head()
```

```python
paris_listings_over_time = paris_listings.groupby("year_of_listing").agg({"listing_id":"count","price":"mean"})
```

```python
paris_listings_over_time
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(data=paris_listings_over_time["listing_id"],color="#00A699")
plt.grid()
plt.title("Listings Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Listings")
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(data=paris_listings_over_time["price"],color="#CE6D47")
plt.title("Price Over Time")
plt.xlabel("Year")
plt.ylabel("Price")
plt.grid()
```

```python

fig,ax=plt.subplots(figsize=(20,8))
line1,=ax.plot(paris_listings_over_time.index,paris_listings_over_time["listing_id"],color="#00A699")
ax2=ax.twinx()
line2,=ax2.plot(paris_listings_over_time.index,paris_listings_over_time["price"],color="#CE6D47")
ax2.set_ylim(0)
lines=[line1,line2]
labels=["Listings","Average Price"]
ax.legend(lines,labels,loc="upper left")
plt.title("PARIS: Average Price Over Time")
ax.set_xlabel("Year")
ax2.set_ylabel("Average Price(€)")
plt.show()
```
